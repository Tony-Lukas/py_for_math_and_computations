The order of operations, or operator precendence, for arithmetic expressions is as follows:

1. Expressions between parentheses
2. Exponents
3. Multiplication and division (from left to right)
4. Addition and subtraction (from left to right)

$$
\text{area} = \sqrt{s(s-a)(s-b)(s-c)} \> \text{where}\> s = \frac{a+b+c}{2}
$$
if `a=5, b=4, c=3` find area of the triangle?

```python
import math
a = 5
b = 4
c = 3
s = (a+b+c)/2
area = math.sqrt(s*(s-a)*(s-b)*(s-c))
print(area)
```
