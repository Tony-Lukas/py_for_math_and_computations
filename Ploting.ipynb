```python
import numpy as np
x = np.linspace(0,5,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,2*np.pi,50)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x,y1,'b.-',label='sin')
ax.plot(x,y2,'r.--',label='cos')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1 , 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# main figure
axes1.plot(x,y,'b')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

# insert figure
axes2.plot(y,x,'r')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0,5,20)
y = np.exp(x)

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x,y)
ax[0].set_title('exponential')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].set_yscale('log')

ax[1].plot(x,y)
ax[1].set_title('logarithmic')
ax[1].set_xlabel('x')
ax[1].set_ylabel('log(y)')

plt.tight_layout()
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = x**2
fig,ax = plt.subplots(2,2)
ax[0,0].bar(x,y)
ax[0,0].set_title('bar')
ax[0,1].scatter(x,y)
ax[0,1].set_title('scatter')
ax[1,0].step(x,y)
ax[1,0].set_title('step')
ax[1,1].plot(x,y)
ax[1,1].set_title('plot')
plt.tight_layout()
plt.show()
```

# 3D plot
$$z=f(x,y) = 100x(1-x)y(1-y)\tan^{-1} \left( 10 \left(\frac{x+y}{\sqrt{2}}-0.8 \right) \right) $$

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,1,20)
y = np.linspace(0,1,20)
X,Y = np.meshgrid(x,y)
Z = 100*X*(1-X)*Y*(1-Y)*np.arctan(10*((X+Y)/np.sqrt(2)-0.8))
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
fig = plt.pcolor(Z)
plt.show()
```

```python
fig = plt.contour(Z,20)
plt.show()
```
